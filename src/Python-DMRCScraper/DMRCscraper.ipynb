```python
# wikipedia is a static website hence i will bs4 to scrape the data from the website
# website link - https://en.wikipedia.org/wiki/List_of_Delhi_Metro_stations
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Delhi_Metro_stations'
req = requests.get(url).text
soup = BeautifulSoup(req, 'lxml')
```

```python
rows = soup.find_all('tr')
rows
```

```python
lines = [
    "Yellow Line",
    "Blue Line",
    "Pink Line",
    "Red Line",
    "Green Line",
    "Violet Line",
    "Magenta Line",
    "Grey Line",
    "Airport Express",
]

station = []
lines = []
```

```python
# obtain a list of stations without the intersection point
for row in rows:
    # Find all <td> elements in the row
    cells = row.find_all('td', align='center')

    # Extract the content of the first and second <td> elements
    if len(cells) >= 2:
        if cells[0].text.strip() in lines:
            continue
        station_name = cells[0].text.strip().replace("*", "").replace("†", "").replace("¤", "").replace("#", "")
        line_name = cells[1].text.strip()

        station.append(station_name)
        lines.append(line_name)
```

```python
for i in range(25, len(station)):
    print(f"{i+1} . {station[i]} - {lines[i]}")
```
